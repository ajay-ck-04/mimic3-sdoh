```python
## Functions

# Tokenize
def tokenize(text):
  doc = nlp(text)
  tokens = [token.text for token in doc if not token.is_punct | token.is_space]
  return tokens

# Lemmatize
def lemmatize(text):
  if isinstance(text,list):
    text = ' '.join(text)
  doc = nlp(text)
  lemmas = ' '.join([token.lemma_ for token in doc])
  return lemmas

# Split
def split(text):
  return text.split()
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(clean2000df['LEMMATIZED'])
```

```python
# Sparse matrix is mostly zeroes; desnse matrix might be helpful as well
dense_matrix = tfidf_matrix.todense()
```

```python
# Finding the most important words

# Sum of TF-IDF scores for each word across all documents
sum_scores = tfidf_matrix.sum(axis=0)

# Convert the sum scores to a dense matrix and get words
word_scores = [(word, sum_scores[0, idx]) for word, idx in tfidf.vocabulary_.items()]

# Sort words by their TF-IDF scores in descending order
word_scores_sorted = sorted(word_scores, key=lambda x: x[1], reverse=True)
```

```python
# Print top x words with highest TF-IDF scores
print("Top 5 words by TF-IDF score:")
for word, score in word_scores_sorted[:5]:
    print(f"{word}: {score:.3f}")
```

```python
# Using spaCy embeddings to define a target theme with document scores
# Definte target theme
target_words = ['bus','homeless', 'commute']
target_embeddings = [nlp(word).vector for word in target_words]
target_centroid = np.mean(target_embeddings, axis=0)

documents = list_of_docs # Set documents

# Compute document embeddings and cosine simularity to target theme
document_scores = []
for doc in documents:
    doc_embedding = np.mean([token.vector for token in nlp(doc)], axis=0)
    similarity = np.dot(doc_embedding, target_centroid) / (np.linalg.norm(doc_embedding) * np.linalg.norm(target_centroid))
    document_scores.append(similarity)
```

```python
# Print top results
top_results_spacy = np.argsort(document_scores)[::-1][:10]
for i, idx in enumerate(top_results_spacy):
    print("Result {}: Document '{}' with cosine similarity score {}".format(i+1, documents[idx], document_scores[idx]))
```
